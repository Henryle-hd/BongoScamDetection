```python
import pandas as pd
```

```python
data=pd.read_csv('scam_trust.csv')
```

```python
data.head()
```

```python
data['scam']=data['Category'].apply(lambda x: 1 if x=='scam' else 0)
data=data.drop(['Category'],axis=1)
```

```python
data.tail(20)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.Sms,data.scam,test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c=CountVectorizer()
X_train_count=c.fit_transform(X_train)
X_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
X_test_count=c.transform(X_test)
model.score(X_test_count,y_test)
```

```python
sms=[
    "wazima uko",
    "Nitumie iyo hela kwenye namba hii 067739272 jina ni KASILU MAUMAU"
]
sms_count=c.transform(sms)
model.predict(sms_count)
```

```python
from sklearn.pipeline import Pipeline
clf=Pipeline([
    ('vectorizer',CountVectorizer()),
    ('np',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(sms)
```

```python
import joblib
```

```python
joblib.dump(clf,'bongo_scam_predict.pkl')
```

```python
loaded_model=joblib.load('bongo_scam_predict.pkl')
pred=loaded_model.predict(['niaje kaka'])
result='Uwizi kaka' if pred[0]==1 else 'SMS Safi iyo'
print(f'Prediction: {result}')
```
